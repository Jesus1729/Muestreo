```python
import random
import statistics
import matplotlib.pyplot as plt
```

```python
datos = 1000
lista_unif = []
for _ in range(datos):
    lista_unif.append(random.uniform(0,1))
```

```python
plt.hist(lista_unif, bins=20, edgecolor='grey',color='purple')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n Uniforme')
plt.show()
```

```python
num_muestras = 1000
lista_muestras = []
for _ in range(num_muestras):
    lista_muestras.append(random.sample(lista_unif, 30))
```

```python
promedios = []
for i in range(len(lista_muestras)):
    promedios.append(statistics.mean(lista_muestras[i]))
```

```python
plt.hist(promedios, bins=20, edgecolor='grey',color='purple') 
plt.xlabel('Promedios') 
plt.ylabel('Frecuencia')
plt.title('Histograma')
plt.show()
```
